```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import font_manager

# Pretendard 폰트를 시스템에서 로드합니다.
font_path = "C:/Users/alaru/AppData/Local/Microsoft/Windows/Fonts/Pretendard-Regular.otf"  # Pretendard 폰트 경로
font_manager.fontManager.addfont(font_path)
plt.rcParams['font.family'] = 'Pretendard'

# 데이터 준비
data = {
    "Tier": ["Bronze", "Silver", "Gold", "Platinum", "Emerald", "Diamond", "C/G/M", "All"],
    "MSE": [0.0626, 0.0588, 0.0557, 0.0550, 0.0501, 0.0494, 0.0453, 0.0529],
    "R^2": [0.7496, 0.7650, 0.7770, 0.7799, 0.7996, 0.8022, 0.8190, 0.7886]
}

df = pd.DataFrame(data)

# 막대그래프 색상 설정
bar_colors = ['#CDFAFA'] * 7 + ['#0AC8B9']  # All에 대해 다른 색상 적용

# 시각화
fig, ax1 = plt.subplots(figsize=(10, 6))

# 배경색 설정
fig.patch.set_facecolor('#091428')
ax1.set_facecolor('#091428')

# R^2를 막대그래프로
bars = ax1.bar(df["Tier"], df["R^2"], color=bar_colors, label="R^2")
# ax1.set_ylabel("R^2", color='white', fontsize=20)  # y축 레이블
ax1.tick_params(axis='y', labelcolor='white', labelsize=16)  # y축 틱 레이블 색상 및 크기
ax1.tick_params(axis='x', colors='white', labelsize=14)  # x축 틱 레이블 색상 및 크기

# 바 그래프 안에 R^2 값을 표시
for bar, r2 in zip(bars, df["R^2"]):
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, yval - 0.05, f'{r2:.2f}', ha='center', va='center', color='black', fontsize=14)

# MSE를 라인그래프로
ax2 = ax1.twinx()
line, = ax2.plot(df["Tier"], df["MSE"], color='#C8AA6E', marker='o', markersize=10, label="MSE")
# ax2.set_ylabel("MSE", color='#C8AA6E', fontsize=16)  # y축 레이블
ax2.tick_params(axis='y', labelcolor='#C8AA6E', labelsize=16)  # y축 틱 레이블 색상 및 크기

# 가운데 점선 제거
ax1.grid(False)
ax2.grid(False)

# 범례 추가 - 차트 바깥 오른쪽 위로 완전히 이동
lines = [bars, line]
labels = [bars.get_label(), line.get_label()]
legend = ax1.legend(lines, labels, loc='center left', bbox_to_anchor=(0.95, 1.1), facecolor='#091428', fontsize=14, edgecolor='white', labelcolor='white')

# 차트 박스 테두리 설정
ax1.spines['top'].set_visible(False)  # 위쪽 테두리 제거
ax1.spines['right'].set_color('white')  # 오른쪽 테두리 색상 흰색
ax1.spines['left'].set_color('white')  # 왼쪽 테두리 색상 흰색
ax1.spines['bottom'].set_color('white')  # 바닥 테두리 색상 흰색

ax2.spines['top'].set_visible(False)  # 위쪽 테두리 제거
ax2.spines['right'].set_color('white')  # 오른쪽 테두리 색상 흰색
ax2.spines['left'].set_color('white')  # 왼쪽 테두리 색상 흰색
ax2.spines['bottom'].set_color('white')  # 바닥 테두리 색상 흰색

# 제목 및 레이블
# plt.title("Comparison of R^2 and MSE by Tier", fontsize=20, color='white')  # 제목
fig.tight_layout()
plt.show()
```

```python
import matplotlib.font_manager as font_manager
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 데이터 불러오기
df = pd.read_csv('LOL_match data_Combined.csv')

# Pretendard 폰트를 시스템에서 로드합니다.
font_path = "C:/Users/alaru/AppData/Local/Microsoft/Windows/Fonts/Pretendard-Regular.otf"  # Pretendard 폰트 경로
font_manager.fontManager.addfont(font_path)
plt.rcParams['font.family'] = 'Pretendard'

# 원하는 순서로 카테고리형 데이터 생성
desired_order = ['Bronze', 'Silver', 'Gold', 'Platinum', 'Emerald', 'Diamond', 'C+GM+M']
df['Tier'] = pd.Categorical(df['Tier'], categories=desired_order, ordered=True)

# Tier별로 그룹화하고 'Value' 컬럼에 대한 카운트 계산
tier_counts = df.groupby('Tier').count()['matchId'].reset_index()

# 그래프 크기 조정
fig, ax = plt.subplots(figsize=(8, 6))  # 가로로 넓게 설정

# 배경색과 차트 배경색 설정
fig.patch.set_facecolor('#091428')
ax.set_facecolor('#091428')

# 각 막대의 색깔 지정
custom_palette = ['#CDFAFA'] * len(desired_order)  # 막대의 색상을 통일

# seaborn 막대차트 그리기 (막대의 폭 조정)
sns.barplot(x='Tier', y='matchId', data=tier_counts, ax=ax, palette=custom_palette, width=0.5)  # 막대의 폭을 0.5로 설정

# x축 글씨 90도 기울이기와 글씨 크기 및 자간 조절
plt.xticks(rotation=0, color='white', fontsize=12)  # 글씨 크기를 12로 설정
plt.gcf().subplots_adjust(bottom=0.25)  # 자간 조정

# y축 글씨 색상 설정
plt.yticks(color='white')

# x축 라벨 제거
plt.xlabel('')

# y축 라벨 색상 설정
# plt.ylabel('Count', color='white')

# 스파인 설정 (테두리 조정)
ax.spines['left'].set_color('white')
ax.spines['bottom'].set_color('white')
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

# plt.title('Count of Values by Tier', color='white')

plt.show()
```

```python
import matplotlib.font_manager as font_manager
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Pretendard 폰트를 시스템에서 로드합니다.
font_path = "C:/Users/alaru/AppData/Local/Microsoft/Windows/Fonts/Pretendard-Regular.otf"  # Pretendard 폰트 경로
font_manager.fontManager.addfont(font_path)
plt.rcParams['font.family'] = 'Pretendard'

# gameType에서 season 추출
df['season'] = df['gameType'].str[:2]

# season별 개수 세기
season_counts = df['season'].value_counts()

# 결과를 데이터프레임으로 변환
result_df = pd.DataFrame({
    'Season': season_counts.index,
    'Count': season_counts.values
})

# 그래프 크기 조정
fig, ax = plt.subplots(figsize=(8, 8))

# 배경색과 차트 배경색 설정
fig.patch.set_facecolor('#091428')
ax.set_facecolor('#091428')

# 도넛 차트 그리기
colors = ['#CDFAFA', '#FF9999']
wedges, texts, autotexts = ax.pie(result_df['Count'], labels=result_df['Season'], colors=colors, 
                                  autopct='%1.1f%%', startangle=90, counterclock=False, 
                                  wedgeprops=dict(width=0.3), textprops=dict(color="white"))

# 글씨 크기 설정
for text in texts:
    text.set_fontsize(14)
for autotext in autotexts:
    autotext.set_fontsize(14)

# 가운데 원 그리기 (도넛 차트 효과)
centre_circle = plt.Circle((0, 0), 0.70, color='#091428', fc='#091428')
fig.gca().add_artist(centre_circle)

# 제목 설정 (선택 사항)
plt.title('Season Counts', color='white', fontsize=16)

# 차트 표시
plt.show()
```
